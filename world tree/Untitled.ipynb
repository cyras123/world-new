```python
import json
```

```python
with open("world-country.json","r") as f:
    country=json.load(f)
```

```python
for key,value in country.items():
    region = key
    subregion = value
    #print(subregion)
    for key,value in subregion.items():
        i=key
        j=value
        print(i)
```

```python
Data = []
i = 1; j = 1; k = 1; l = 1
Data.append({"id" : "World", "value":str(i)})

for key,value in country.items():
    region = key
    subregion = value
    
    Data.append({"id":"World"+"."+region,\
                 "value":str(i) + "." + str(j)})
    
    j =j+1
    
    for key,value in subregion.items():
        
        subcontinent = key
        country = value
        #print(country)
        
        Data.append({"id":"World"+"."+region+"."+subcontinent,\
                     "value":str(i) + "." + str(j) + "." +str(k)})
        k = k+1
        
        
        for cnty in country:
            #print(i)
            mcp = cnty
            l = l+1
            
            Data.append({"id":"World"+"."+region+"."+subcontinent+"."+ mcp,\
                         "value":str(i) + "." + str(j) + "." +str(k) + "."+str(k)})
```
